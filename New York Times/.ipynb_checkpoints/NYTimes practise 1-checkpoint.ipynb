```python
import requests
from bs4 import BeautifulSoup
```

```python
nytimes = 'https://www.nytimes.com/books/best-sellers/combined-print-and-e-book-fiction/'
```

```python
nytimes_req = requests.get(nytimes).text
```

```python
nytimes_req
```

```python
nytimes_soup = BeautifulSoup(nytimes_req, 'lxml')
nytimes_soup
```

```python
soup_select = nytimes_soup.select('article')[0]
soup_select
```

```python
soup_select = nytimes_soup.select('article h3')
```

```python
soup_select
```

```python
result = nytimes_soup.select('article h3[itemprop="name"]')
```

```python
[x.text for x in result]
```

```python
soup_select = nytimes_soup.select('article')[0]
soup_select
```

```python
summ = nytimes_soup.select('article p[itemprop="description"]')
summ
```

```python
for x in summ:
    print(x.text)
```

```python
soup_select = nytimes_soup.select('article p[itemprop="author"]')
soup_select
```

```python
for x in soup_select:
    print(x.text.strip('by'))
```
